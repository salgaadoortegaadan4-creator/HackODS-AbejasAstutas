### Análisis de la Brecha Digital: Suroeste vs. Nacional

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de conectividad (ENDUTIH 2023)
datos = {
    'Estado': ['Promedio Nacional', 'Chiapas', 'Oaxaca', 'Guerrero'],
    'Hogares con Internet (%)': [73.6, 50.7, 55.5, 58.9]
}

df = pd.DataFrame(datos)

# Creación de la visualización
plt.figure(figsize=(10, 6))
colores = ['#2ecc71', '#e74c3c', '#f39c12', '#f39c12']
plt.bar(df['Estado'], df['Hogares con Internet (%)'], color=colores)
plt.axhline(y=73.6, color='blue', linestyle='--', label='Meta Nacional')

plt.title('Disponibilidad de Internet en el Suroeste de México')
plt.ylabel('Porcentaje de Hogares (%)')
plt.ylim(0, 100)
plt.legend()
plt.show()
```

**Conclusión del análisis:** Se observa un rezago histórico en Chiapas, donde la conectividad es casi 23 puntos menor al promedio nacional, afectando directamente el ODS 4 (Educación).
