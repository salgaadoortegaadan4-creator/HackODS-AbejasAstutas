## <strong>Introducción</strong>
</div>

El documento que estás leyendo no es una página web estática, sino un entorno interactivo denominado <strong>notebook de Colab</strong>, que permite escribir y ejecutar código.

Por ejemplo, esta es una <strong>celda de código</strong> con una secuencia de comandos Python corta que calcula un valor, lo almacena en una variable y devuelve el resultado:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Las variables que defines en una celda pueden usarse en otras:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

## Ciencia de datos
</div>

Con Colab, puedes aprovechar por completo las bibliotecas más populares de Python para analizar y visualizar datos. La celda de código que se incluye a continuación usa <strong>NumPy</strong> para generar algunos datos aleatorios y <strong>matplotlib</strong> para visualizarlos. Para editar el código, haz clic en la celda y comienza a editar.

Puedes importar datos propios a notebooks de Colab desde tu cuenta de Google Drive &#40;incluso desde hojas de cálculos&#41;, GitHub y muchas otras fuentes. Para obtener más información acerca de la importación de datos y cómo puede usarse Colab para fines relacionados con la ciencia de datos, consulta los vínculos de <a href="#working-with-data">Cómo trabajar con datos</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap
from PIL import Image
import requests
from io import BytesIO

# --- CONFIGURACIÓN DE PANTALLA ---
plt.rcParams['figure.facecolor'] = '#fdfdfd'
fig = plt.figure(figsize=(24, 28)) # Lienzo extra alto para evitar amontonamiento
# 4 filas: Título, Historia, Mapa/Tecnología, Pobreza/Educación
gs = fig.add_gridspec(4, 2, height_ratios=[0.2, 0.4, 1.2, 1], hspace=0.3, wspace=0.15)

# --- ENCABEZADO ---
fig.suptitle('DESCONECTADOS: EL COSTO DE ESTUDIAR EN EL SUROESTE\nEquipo: AbejasAstutas | HackODS 2026',
             fontsize=38, fontweight='bold', color='#154360', y=0.98)

# --- ZONA 1: LA HISTORIA (Storytelling fluido) ---
ax_text = fig.add_subplot(gs[1, :])
ax_text.axis('off')
texto_prosa = (
    "La educación en el siglo XXI no se mide solo en libros o pupitres, sino en latencia y ancho de banda. "
    "Para miles de estudiantes en el rincón sur de nuestro país, el futuro se desvanece cada vez que el cielo se nubla. "
    "En regiones como Chiapas, Guerrero y Oaxaca, el rezago social define el punto de partida. Mientras el 73.6% de los hogares "
    "en México tiene internet, Chiapas apenas alcanza el 50.7%. Esta brecha no se soluciona solo repartiendo tabletas; "
    "un equipo moderno sin red es solo un trozo de plástico. La infraestructura resiliente es una deuda histórica que "
    "sigue perpetuando el ciclo de exclusión. Sin señal, el futuro se pone en pausa."
)
wrap_text = textwrap.fill(texto_prosa, width=120)
ax_text.text(0.5, 0.5, wrap_text, fontsize=21, ha='center', va='center', linespacing=1.8,
             bbox=dict(facecolor='#f4faff', alpha=1, edgecolor='#3498db', boxstyle='round,pad=2'))

# --- ZONA 2: EL MAPA (Grande y claro) ---
ax_map = fig.add_subplot(gs[2, 0])
url_mapa = "https://www.coneval.org.mx/Medicion/MP/Documents/Pobreza_Municipal_2020/Mapa_Pobreza_Municipal_2020.jpg"
try:
    response = requests.get(url_mapa, timeout=10)
    img = Image.open(BytesIO(response.content))
    ax_map.imshow(img)
    ax_map.set_title('MAPA DE POBREZA MUNICIPAL (CONEVAL 2020)', fontsize=22, fontweight='bold', pad=20)
    ax_map.axis('off')
except:
    ax_map.text(0.5, 0.5, 'Sube el archivo "mapa.jpg" a Colab\ny usa: img = Image.open("mapa.jpg")',
                ha='center', va='center', fontsize=18, color='red')
    ax_map.axis('off')

# --- ZONA 3: BRECHA DIGITAL (ENDUTIH 2024) ---
ax_tech = fig.add_subplot(gs[2, 1])
df_tech = pd.DataFrame({
    'Estado': ['Nacional', 'Chiapas', 'Oaxaca', 'Guerrero'],
    'Internet (%)': [73.6, 50.7, 55.5, 58.9],
    'Disponibilidad PC (%)': [43.8, 19.4, 25.1, 28.3]
})
df_tech.plot(x='Estado', kind='bar', ax=ax_tech, color=['#2980b9', '#aed6f1'], width=0.7)
ax_tech.set_title('ACCESO A TECNOLOGÍA EN HOGARES', fontsize=22, fontweight='bold', pad=20)
ax_tech.set_ylim(0, 100)
ax_tech.legend(fontsize=16)
ax_tech.tick_params(labelsize=14)

# --- ZONA 4: POBREZA (CONEVAL 2023) ---
ax_pob = fig.add_subplot(gs[3, 0])
sns.barplot(x=['Nacional', 'Chiapas', 'Oaxaca', 'Guerrero'], y=[36.3, 67.4, 58.4, 60.4],
            palette='Reds_r', ax=ax_pob)
ax_pob.set_title('POBREZA MULTIDIMENSIONAL (%)', fontsize=22, fontweight='bold', pad=20)
ax_pob.set_ylabel('Porcentaje (%)', fontsize=16)
ax_pob.tick_params(labelsize=14)

# --- ZONA 5: EDUCACIÓN (SIGED/SEP) ---
ax_edu = fig.add_subplot(gs[3, 1])
niveles = ['Primaria', 'Secundaria', 'Media Sup.', 'Superior']
nac = [98.5, 92.1, 75.3, 45.2]
sur = [94.2, 85.0, 58.1, 22.4]
ax_edu.plot(niveles, nac, marker='s', markersize=12, linewidth=5, label='Nacional', color='#27ae60')
ax_edu.plot(niveles, sur, marker='o', markersize=12, linewidth=5, label='Suroeste', color='#c0392b')
ax_edu.set_title('TASA DE FINALIZACIÓN DE ESTUDIOS', fontsize=22, fontweight='bold', pad=20)
ax_edu.set_ylim(0, 110)
ax_edu.legend(fontsize=16)
ax_edu.grid(True, alpha=0.3)
ax_edu.tick_params(labelsize=14)

# PREGUNTA FINAL
pregunta = "¿Debería el internet ser legalmente un servicio público básico para reducir esta brecha?"
fig.text(0.5, 0.02, pregunta, fontsize=24, fontweight='bold', ha='center',
         bbox=dict(facecolor='#fef9e7', edgecolor='#f1c40f', boxstyle='round,pad=1'))

plt.show()
```
